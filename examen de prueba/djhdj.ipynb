*Ejercicio 1*: El Validador de Contraseñas "Paranoico" 🔒
Crear un simple validador de contraseñas es fácil. Vamos a crear uno que tenga reglas complejas y contradictorias, simulando un sistema de alta seguridad.

El Reto
Escribe una función validar_password(password) que reciba un string. La función debe devolver un string indicando si la contraseña es "Aceptada" o un mensaje específico de por qué fue "Rechazada".

Las Reglas (Deben cumplirse TODAS)
Longitud: Debe tener entre 12 y 20 caracteres.

Caracteres Obligatorios:

Debe contener al menos dos números (0-9).

Debe contener al menos una letra mayúscula y una letra minúscula.

Debe contener al menos un carácter especial (ej. !, @, #, $, %, &, *).

Caracteres Prohibidos:

No puede contener espacios en blanco.

No puede contener la secuencia "123" ni la secuencia "abc" (ni mayúsculas "ABC").

La Regla "Paranoica" (El Twist):

Si la contraseña tiene exactamente 13 caracteres, NO puede empezar con un número.

Si la contraseña tiene más de 18 caracteres, DEBE terminar con un carácter especial

*Ejercicio 2*: Sistema de Triage de Emergencias 🚑
En un hospital, el triage clasifica a los pacientes por urgencia. Vamos a simular este sistema basándonos en signos vitales. El orden de las comprobaciones es crítico.

El Reto
Escribe una función asignar_triage(signos_vitales) que recibe un diccionario. La función debe devolver el color de prioridad: "ROJO (Crítico)", "NARANJA (Urgente)", "AMARILLO (Moderado)", o "VERDE (Leve)".

Ejemplo de entrada: paciente1 = {"consciente": True, "presion_sistolica": 110, "frecuencia_respiratoria": 18, "saturacion_oxigeno": 95}

Las Reglas (Se evalúan en este orden estricto)
Prioridad ROJO (Crítico):

Si el paciente está inconsciente (consciente: False).

O si la saturación de oxígeno es menor a 90%.

O si la frecuencia respiratoria es menor a 10 O mayor a 30.

Prioridad NARANJA (Urgente):

(Si no es ROJO) Y si la presión sistólica es mayor a 180 O menor a 90.

O si la saturación de oxígeno está entre 90% y 93% (inclusive).

Prioridad AMARILLO (Moderado):

(Si no es ROJO ni NARANJA) Y si el paciente está consciente PERO reporta dolor severo (supongamos que es una variable dolor_severo: True en el diccionario, si existe).

O si la presión sistólica está entre 140 y 180.

O si la frecuencia respiratoria está entre 25 y 30.

Prioridad VERDE (Leve):

Si no cumple ninguno de los criterios anteriores.

*Ejercicio 3*: Calculadora de Impuestos Compuestos 💰
    Olvídate de un simple impuesto sobre la renta. Vamos a calcular un impuesto que depende de la categoría del producto, el precio base y el estatus del comprador.

    El Reto
Escribe una función calcular_impuesto(precio, categoria, es_miembro) que recibe un número (precio), un string (categoría) y un booleano (es_miembro). Debe devolver el precio total (precio + impuestos).

    Las Reglas del Impuesto
Impuesto Base: Todos los productos tienen un impuesto base del 10%.

    Impuestos de Categoría (Acumulativos):

Si la categoría es "Electronica": se añade un 8% extra.

Si la categoría es "Lujo": se añade un 20% extra.

Si la categoría es "Alimentos": el impuesto base se reduce al 3% (en lugar del 10%).

    Impuesto de Rango de Precio (El Twist):

Este impuesto se aplica después de los impuestos de categoría.

Si el precio (original) es mayor a $500 pero menor o igual a $2000: se añade un 5% fijo (calculado sobre el precio original).

Si el precio es mayor a $2000: se añade un 10% fijo (calculado sobre el precio original).

    Descuento de Membresía:

Si es_miembro es True, se aplica un descuento del 5% sobre el total de impuestos calculados (no sobre el precio base).

Excepción: Los miembros no reciben descuento en la categoría "Lujo".
